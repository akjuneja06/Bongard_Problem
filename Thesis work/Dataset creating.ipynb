```python
# Function to generate a random shape
def generate_shape(image_size, num_shapes, filled_n):
    if filled_n==1:
        filled = 0
        unfill = 255
    else:
        filled = 255
        unfill = 0
        
    image = Image.new("L", image_size, filled)  # Create a blank white image
    draw = ImageDraw.Draw(image)

    center_x = image_size[0] // 2
    center_y = image_size[1] // 2
    shape_size = random.randint(10, 20)

    # Generate the shapes in the center and not in the center
    shape_types = ["circle", "square", "triangle", "pentagon", "hexagon", "octagon", "star",
                   "rhombus", "cross", "arrow", "heart", "crescent", "trapezoid", "parallelogram", "diamond"]

    center_shape = random.choice(shape_types)
    shape_types.remove(center_shape)

    # Generate the shape in the center
    if num_shapes > 1:
        draw_shape(center_shape, draw, center_x, center_y, shape_size, unfill)

    # Generate the other shapes not in the center
    for _ in range(num_shapes - 1):
        shape_type = random.choice(shape_types)
        shape_x = random.randint(0, image_size[0] - shape_size)
        shape_y = random.randint(0, image_size[1] - shape_size)
        draw_shape(shape_type, draw, shape_x, shape_y, shape_size, unfill)

    return image

# Function to draw different shapes
def draw_shape(shape_type, draw, x, y, size, fill):
    if shape_type == "circle":
        draw.ellipse((x, y, x + size, y + size), fill=fill)
    elif shape_type == "square":
        draw.rectangle((x, y, x + size, y + size), fill=fill)
    elif shape_type == "triangle":
        points = [(x + size // 2, y), (x, y + size), (x + size, y + size)]
        draw.polygon(points, fill=fill)
    elif shape_type == "pentagon":
        angle = 360 / 5
        points = [(x + size * math.cos(math.radians(angle * i)), y + size * math.sin(math.radians(angle * i))) for i in range(5)]
        draw.polygon(points, fill=fill)
    elif shape_type == "hexagon":
        angle = 360 / 6
        points = [(x + size * math.cos(math.radians(angle * i)), y + size * math.sin(math.radians(angle * i))) for i in range(6)]
        draw.polygon(points, fill=fill)
    elif shape_type == "octagon":
        angle_offset = 22.5
        points = [(x + size * math.cos(math.radians(45 + angle_offset * i)), y + size * math.sin(math.radians(45 + angle_offset * i))) for i in range(8)]
        draw.polygon(points, fill=fill)
    elif shape_type == "star":
        
        num_points = 5
        outer_radius = size
        inner_radius = size // 2

        outer_points = [
            (
                x + outer_radius * math.cos(2 * math.pi * i / num_points),
                y + outer_radius * math.sin(2 * math.pi * i / num_points),
            )
            for i in range(num_points)
        ]
        inner_points = [
            (
                x + inner_radius * math.cos(2 * math.pi * (i + 0.2) / num_points),
                y + inner_radius * math.sin(2 * math.pi * (i + 0.2) / num_points),
            )
            for i in range(num_points)
        ]
        all_points = [point for pair in zip(outer_points, inner_points) for point in pair]
        draw.polygon(all_points, fill=fill)

    elif shape_type == "rhombus":
        points = [(x + size // 2, y), (x, y + size // 2), (x + size // 2, y + size), (x + size, y + size // 2)]
        draw.polygon(points, fill=fill)
    elif shape_type == "cross":
        draw.line((x, y + size // 2, x + size, y + size // 2), fill=fill)
        draw.line((x + size // 2, y, x + size // 2, y + size), fill=fill)
    elif shape_type == "arrow":
        points = [(x + size // 2, y), (x, y + size), (x + size // 2, y + size // 2), (x + size, y + size)]
        draw.polygon(points, fill=fill)
    elif shape_type == "heart":
        top_curve = [(x + size * math.cos(math.radians(i)), y - size * math.sin(math.radians(i))) for i in range(135, 316)]
        bottom_curve = [(x - size * math.cos(math.radians(i)), y + size * math.sin(math.radians(i))) for i in range(225, 46, -1)]
        all_points = top_curve + bottom_curve
        draw.polygon(all_points, fill=fill)
    elif shape_type == "crescent":
        num_points = 16
        outer_points = [(x + size * math.cos(2 * math.pi * i / num_points), y + size * math.sin(2 * math.pi * i / num_points)) for i in range(num_points)]
        inner_points = [(x + (size // 1.5) * math.cos(2 * math.pi * i / num_points), y + (size // 1.5) * math.sin(2 * math.pi * i / num_points)) for i in range(num_points)]
        all_points = [point for pair in zip(outer_points, inner_points) for point in pair]
        draw.polygon(all_points, fill=fill)

    elif shape_type == "trapezoid":
        top_base = size * 1.5
        bottom_base = size * 3
        height = size * 2
        points = [(x - bottom_base // 2, y + height // 2), (x + bottom_base // 2, y + height // 2),
                  (x + top_base // 2, y - height // 2), (x - top_base // 2, y - height // 2)]
        draw.polygon(points, fill=fill)
    elif shape_type == "parallelogram":
        top_base = size * 2
        height = size * 2
        points = [(x - top_base // 2, y + height // 2), (x + top_base // 2, y + height // 2),
                  (x + (top_base // 2) - size, y - height // 2), (x - (top_base // 2) - size, y - height // 2)]
        draw.polygon(points, fill=fill)
    elif shape_type == "diamond":
        points = [(x + size // 2, y), (x, y + size // 2), (x + size // 2, y + size), (x + size, y + size // 2)]
        draw.polygon(points, fill=fill)
        

# Number of images to generate
num_images = 200
# Number of shapes in each image
n = 2  # Change this to the desired number of shapes in each image
# Output directory
output_dir = "C:/python39/Dataset/Foundalis_Bongard_images/Classifiers I generated/center  vs not incenter/Generated"

generated_images = 0
while generated_images < num_images:
    num_shapes = n
    image = generate_shape((100, 100), num_shapes,2)
    if image.getextrema() != (255, 255):  # Check if the image is not blank white
        image.save(os.path.join(output_dir, f"C{generated_images+1}.png"))
        generated_images += 1
        
print("All done white image")

# Number of images to generate
num_images = 200
# Number of shapes in each image
n = 2  # Change this to the desired number of shapes in each image
# Output directory
output_dir = "C:/python39/Dataset/Foundalis_Bongard_images/Classifiers I generated/center  vs not incenter/Generated"
generated_images = 0
while generated_images < num_images:
    num_shapes = n
    image = generate_shape((100, 100), num_shapes,1)
    if image.getextrema() != (255, 255):  # Check if the image is not blank white
        image.save(os.path.join(output_dir, f"D{generated_images+1}.png"))
        generated_images += 1
        
print("All done black image")
```
